```python
# Animal Crossing Data Analyst Portfolio Project
# Item Price Analysis
# By Donya Garrison
# Dataset: https://www.kaggle.com/datasets/jessicali9530/animal-crossing-new-horizons-nookplaza-dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the dataset 'accessories.csv' 
DATA_PATH = '/Users/donyagarrison/Desktop/acnh-item-price-analysis/data/raw/accessories.csv'  
df = pd.read_csv(DATA_PATH)

# Display first few rows
display(df.head())
    
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert Sell to numeric, drop rows where Sell is missing or 0
df['sell'] = pd.to_numeric(df['sell'], errors='coerce')
df = df.dropna(subset=['sell'])
df = df[df['sell'] > 0]

# Drop duplicate rows if any
df = df.drop_duplicates()

# Show info and missing values
display(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())

# Optional: Fill NA for categorical features
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('Unknown')

# EDA: Distribution of Sell prices
plt.figure(figsize=(10,6))
sns.histplot(df['sell'], bins=40, kde=True)
plt.title('Distribution of Sell Prices')
plt.xlabel('Sell Price')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Top 15 most valuable items
top_items = df.sort_values('sell', ascending=False).head(15)
display(top_items[['name', 'sell']])

# Categorical Feature Impact (example: category, source, style, color)
categorical_features = [col for col in ['category', 'source', 'style', 'color_1', 'color_2'] if col in df.columns]

for cat in categorical_features:
    plt.figure(figsize=(12,6))
    order = df.groupby(cat)['sell'].median().sort_values(ascending=False).index
    sns.boxplot(x=cat, y='sell', data=df, order=order)
    plt.xticks(rotation=45)
    plt.title(f'Sell Price by {cat.capitalize()}')
    plt.tight_layout()
    plt.show()
    display(df.groupby(cat)['sell'].describe().sort_values('50%', ascending=False).head(8))

# Correlation matrix for numeric features
plt.figure(figsize=(7,5))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Numeric Features)')
plt.tight_layout()
plt.show()

# Feature importance using simple linear regression (optional)
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder

# Select some categorical and numeric features
features = []
for col in ['category', 'source', 'style', 'color_1']:
    if col in df.columns:
        features.append(col)

# Prepare data for regression (drop rows with unknown or rare categories)
X = df[features]
y = df['sell']

# One-hot encoding for categorical features
X_encoded = pd.get_dummies(X, drop_first=True)

# Fit model
model = LinearRegression()
model.fit(X_encoded, y)
importance = pd.Series(model.coef_, index=X_encoded.columns).sort_values(ascending=False)

# Plot top 10 influential features
plt.figure(figsize=(8,5))
importance.head(10).plot(kind='barh')
plt.title('Top 10 Feature Importances (Linear Regression Coefs)')
plt.xlabel('Coefficient')
plt.tight_layout()
plt.show()


# Conclusions
print("""
Conclusions:
- The most valuable items and their features are shown above.
- Categorical features like category, source, and style may influence sell price.
- Feature importances suggest which item attributes contribute most to value.
- Most items have low to moderate sell prices, with a few high-value outliers.
- Further analysis can use non-linear models or focus on specific item types.

## Personal Reflection

Through this project I practiced data cleaning, EDA, and feature importance analysis. I learned how item attributes like source and category affect value. For future work, I'd like to analyze other datasets (fish, bugs) or try to predict prices with machine learning.
""")
```
